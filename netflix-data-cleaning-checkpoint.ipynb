 # Check for missing values
 We identify which columns contain missing data using .isnull().sum().


```python
df.isnull().sum()
```

```python
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Not Available")
df['country'] = df['country'].fillna("Unknown")
df['rating'] = df['rating'].fillna("Unrated")
df['duration'] = df['duration'].fillna("Unknown")
```

# Fill missing values
We fill the missing values with "Unknown" for text columns like director, cast, country, rating, and duration. This ensures that our dataset is consistent and doesn’t break during analysis.

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

# Convert date_added to datetime
The date_added column contains the date a show was added to Netflix. We convert this to a datetime format so we can analyze trends over time. We use errors='coerce' to handle any invalid formats.

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df = df.dropna(subset=['date_added'])
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df['date_added'] = df['date_added'].str.strip()
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
df.isnull().sum()
```

# Netflix Dataset - Exploratory Data Analysis (EDA)

Now that our data is clean, let’s explore it to extract insights and trends. We will use `pandas` and `matplotlib`/`seaborn` for visualizations.

##  Import visualization libraries

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

# Distribution of Content Type (Movies vs TV Shows)
This shows how much of Netflix’s content is Movies vs TV Shows.

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type', hue='type', palette='Set2', legend=False)
plt.title("Distribution of Content Type on Netflix")
plt.xlabel("Type of Content")
plt.ylabel("Count")
plt.show()
```

 # Top 10 Countries with Most Content
This reveals which countries produce the most Netflix content.

```python
# Top countries producing Netflix content
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(8,6))
sns.barplot(
    x=top_countries.values,
    y=top_countries.index,
    hue=top_countries.index,     
    legend=False                 
)
plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

 # Trend of Content Added Over the Years
This helps understand how Netflix's content library has grown year over year.

```python
df['year_added'] = df['date_added'].dt.year
content_per_year = df['year_added'].value_counts().sort_index()

plt.figure(figsize=(10,4))
sns.lineplot(x=content_per_year.index, y=content_per_year.values, marker='o')
plt.title("Number of Titles Added Each Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

# Most Popular Genres (Top 10)
This tells us which genres dominate the platform.

```python
genres = df['listed_in'].str.split(', ').explode().value_counts().head(10).index
counts = df['listed_in'].str.split(', ').explode().value_counts().head(10).values

plt.figure(figsize=(8,6))
sns.barplot(
    x=list(counts),
    y=list(genres),
    hue=list(genres),     
    palette='mako',
    legend=False          
)
plt.title("Top 10 Most Common Genres on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.show()
```

# Top 10 Directors on Netflix
This shows which directors have the most content on Netflix.

```python
# Top 10 directors with the most titles
top_directors = df['director'].value_counts().head(10)

plt.figure(figsize=(8,6))
sns.barplot(
    x=top_directors.values,
    y=top_directors.index,
    hue=top_directors.index,   # assign y as hue
    palette='crest',
    legend=False               # hide legend
)
plt.title("Top 10 Directors with Most Titles")
plt.xlabel("Number of Titles")
plt.ylabel("Director")
plt.show()
```
